## Задание 2
Создайте диагональную матрицу с элементами от N до 0. Посчитайте сумму ее значений на диагонали.

```python
import numpy as np

N = 5

diagonal_elements = np.arange(N, -1, -1)
print("Diagonal elements:", diagonal_elements)


matrix = np.diag(diagonal_elements)
print("Matrix:")
print(matrix)

diagonal_sum = np.trace(matrix)
print("Diagonal summ:", diagonal_sum)
```

## Задание 3
Решите систему уравнений:
* 4x + 2y + z = 4
* x + 3y = 12
* 5y + 4z = -3

```python
# для самостоятельного изучения
import numpy as np
from numpy import linalg

A = np.array([[4, 2, 1], [1, 3, 0], [0, 5, 4]])  # K-matrix
B = np.array([4, 12, -3])  # array right side

X = linalg.solve(A, B)  # ez math


print(f"x = {X[0]:.2f}")  # float 0.00
print(f"y = {X[1]:.2f}")
print(f"z = {X[2]:.2f}")
```

## Задание 7

Создайте две матрицы 2 на 3 и сделайте их горизонтальный стакинг 

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8, 9],
              [10, 11, 12]])

result = np.hstack((A, B))

print(A)
print('---')
print(B)
print('---')
print(result)
print(f"Result: {result.shape}")
```

## Задание 9

Напишите функцию, которая на вход принимает 2 вектора и возвращает тот из них, что больше. Функцию необходимо векторизовать.

```python
import numpy as np

# Функцию необходимо векторизовать. ???

def get_larger_vector(a, b):
    if np.all(a >= b) and np.any(a > b):
        return a
    elif np.all(b >= a) and np.any(b > a):
        return b
    else:
        return a


v1 = np.array([1, 2, 3])
v2 = np.array([1, 2, 4])
v3 = np.array([1, 2, 3])

print("v1:", v1)
print("v2:", v2)
print("v3:", v3)
print("v1 VS v2: v2", get_larger_vector(v1, v2))
print("v1 VS v3: v3", get_larger_vector(v1, v3))
```

## Задание 10

Используя пример из классной работы с рисованием графика среднемесячной температуры, создайте два вектора длиной 100 следуя следующим условиям: 
Вектор 1: случайные числа
Вектор 2: нормальное распределение

Визуализируйте оба вектора.

```python
import matplotlib.pyplot as plt
data = np.genfromtxt("stockholm_td_adj.dat")
data.shape

""" lesson example
fig, ax = plt.subplots(figsize= (14, 4))
ax.plot(data[:,0] + data[:,1]/12.0 + data[:, 2]/365, data[:, 5])
"""

# top-notch sota meta boxplots

np.random.seed(42)

vector1 = np.random.uniform(0, 30, 100)  # 0-30 , 100 items

vector2 = np.random.normal(15, 5, 100)  # normal_distr: mean 15 мат ожидание (среднее), std 5 (ст.отклонение)


time_axis = np.arange(100) / 365.0  # временная ось для 100 дней

print("Array stats:")
print(f"array1 (rnd): mean={np.mean(vector1):.2f}, std={np.std(vector1):.2f}")
print(f"array2 (nrm): mean={np.mean(vector2):.2f}, std={np.std(vector2):.2f}")

fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(time_axis, vector1, 'b-', linewidth=1.5, alpha=0.7, label='Random')
ax.plot(time_axis, vector2, 'r-', linewidth=1.5, alpha=0.7, label='Normal')

ax.set_xlabel('Years')
ax.set_ylabel('temp')
ax.set_title('Comparison Random/Normal vectors')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

# Random
ax1.plot(time_axis, vector1, 'bo-', markersize=3, linewidth=1, alpha=0.7)
ax1.set_ylabel('temp')
ax1.set_title('Random 0-30')
ax1.grid(True, alpha=0.3)

# Normal
ax2.plot(time_axis, vector2, 'ro-', markersize=3, linewidth=1, alpha=0.7)
ax2.set_xlabel('Years')
ax2.set_ylabel('temp')
ax2.set_title('Normal (μ=15, σ=5)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.hist(vector1, bins=15, alpha=0.7, color='blue', edgecolor='black')
ax1.set_title('Random')
ax1.set_xlabel('temp')
ax1.set_ylabel('frequency')

ax2.hist(vector2, bins=15, alpha=0.7, color='red', edgecolor='black')
ax2.set_title('Normal')
ax2.set_xlabel('temp')
ax2.set_ylabel('frequency')

plt.tight_layout()
plt.show()
```
