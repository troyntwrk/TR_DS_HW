#### Описание признаков

**age**: The person's age in years

**sex**: The person's sex (1 = male, 0 = female)

**cp**: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)

**trestbps**: The person's resting blood pressure (mm Hg on admission to the hospital)

**chol**: The person's cholesterol measurement in mg/dl

**fbs**: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)

**restecg**: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)

**thalach**: The person's maximum heart rate achieved

**exang**: Exercise induced angina (1 = yes; 0 = no)

**oldpeak**: ST depression induced by exercise relative to rest ('ST' relates to positions on the ECG plot. See more here)

**slope**: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)

**ca**: The number of major vessels (0-3)

**thal**: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)

**target**: Heart disease (0 = no, 1 = yes)

**Посчитайте у кого уровень депрессии при физический нагрузке выше (в среднем), среди мужчин страдающих болезнью сердца или среди женщин не страдающих болезнью сердца (st_depression)**

**Посчтитайте максимальный и минимальный уровень холестерина для каждого типа chest_pain_type, rest_ecg, thalassemia. Пишите код оптимально, можно использовать циклы**

```python
columns_to_analyze = ['chest_pain_type', 'rest_ecg', 'thalassemia']

for column in columns_to_analyze:
    print(f"\n\nСтатистика холестерина по группам '{column}':")
    
    result = data.groupby(column)['cholesterol'].agg(['min', 'max'])
    print(result)
```

**Сколько значений может принимать каждый из категориальных признаков?**

```python
cat_columns = ['sex', 'chest_pain_type', 'fasting_blood_sugar', 'rest_ecg', 
                       'exercise_induced_angina', 'st_slope', 'thalassemia', 'target']

for col in cat_columns:
    unique_values = data[col].nunique()
    values_list = data[col].unique()
    print(f"{col}:")
    print(f"\tКоличество уникальных значений: {unique_values}")
    print(f"\tЗначения: {sorted(values_list) if isinstance(values_list[0], (int, float)) else values_list}")
    print("\n")
```

**У какого категориального признака наблюдается самый сильный дизбаланс классов?**

```python
cat_columns = ['sex', 'chest_pain_type', 'fasting_blood_sugar', 'rest_ecg', 
                       'exercise_induced_angina', 'st_slope', 'thalassemia', 'target']

max_imbalance = 0
most_imbalanced_col = None
imbalance_info = {}

print("Дисбаланс классов по категориальным признакам:")
print("--------")

for col in cat_columns:
    value_counts = data[col].value_counts(normalize=True)
    
    imbalance_ratio = value_counts.max() / value_counts.min()
    
    imbalance_info[col] = {
        'ratio': imbalance_ratio,
        'max_class': value_counts.idxmax(),
        'max_percent': value_counts.max() * 100,
        'min_class': value_counts.idxmin(),
        'min_percent': value_counts.min() * 100,
        'distribution': value_counts
    }
    
    print(f"\n{col}:")
    print(f"\tСоотношение: {imbalance_ratio:.2f}x") # во сколько раз самый частый класс больше самого редкого класса
    print(f"\tfreq: '{value_counts.idxmax()}' - {value_counts.max():.1%}")
    print(f"\trare: '{value_counts.idxmin()}' - {value_counts.min():.1%}")
    
    if imbalance_ratio > max_imbalance:
        max_imbalance = imbalance_ratio
        most_imbalanced_col = col

print("\n\n")
print(f"Топ1 дисбаланс у класса: {most_imbalanced_col}")
print(f"Соотношение классов: {max_imbalance:.2f}x")
print(f"Распределение:")
for value, percent in imbalance_info[most_imbalanced_col]['distribution'].items():
    print(f"  '{value}': {percent:.1%}")
```
